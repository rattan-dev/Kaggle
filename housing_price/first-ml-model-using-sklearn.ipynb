# Random Forest Regresseor

```python
# importing useful libraries

from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# getting the training data 

data_path = '../input/home-data-for-ml-course/train.csv'
data = pd.read_csv(data_path)
test_data_path = '../input/home-data-for-ml-course/test.csv'
test_data = pd.read_csv(test_data_path)

# target variable
y = data.SalePrice
```

```python
# features
features = ['MSSubClass', 'LotArea', 'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'FullBath', 'HalfBath', 'KitchenAbvGr', 'Fireplaces', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'YrSold']
X = data[features]
```

```python
# splitting the data into train and validation data set
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=3)
# creating a random forest model to train
rf_model = RandomForestRegressor(random_state = 3)
```

```python
# fitting the model
rf_model.fit(train_X, train_y)
```

```python
# validation
validation = rf_model.predict(val_X)

# checking the mean absolute error
mae = mean_absolute_error(validation, val_y)
print(mae)
```

```python
# finally training over whole training data
rf_model.fit(X,y)

# testing
predictions = rf_model.predict(test_data[features])
```

```python
output_rf = pd.DataFrame({'Id': test_data.Id, 'SalePrice': predictions})
```

Done, let's make the submission

# Let's try with decision tree regressor

```python
# importing Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
# once again splitting the data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
```

```python
# determinig the max leaf nodes
candidate = [30, 40, 50, 60, 70, 80]
results = {}
for max_leaf in candidate :
    model = DecisionTreeRegressor(max_leaf_nodes = max_leaf, random_state = 0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(preds_val, val_y)
    results[max_leaf] = mae
 
best_leaf = min(results, key=results.get)    
print(best_leaf)
```

```python
print(results)
```

```python
# creating the final model
final_model = DecisionTreeRegressor(max_leaf_nodes = best_leaf, random_state = 3)
final_model.fit(X, y)

predictions = final_model.predict(test_data[features])
predictions
```

```python
output = pd.DataFrame({'Id': test_data.Id, 'SalePrice': predictions})
```
